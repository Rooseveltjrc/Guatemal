```python
def mayor(datos):
    '''
    Función mayor
    Toma cada una de las variables que se encuentran en el base, para realizar el contéo de las veces que aparece, así generar la frecuencia y seleccionar
    el valor de mayor frecuencia
    repeticiones: espacio donde se guardará el listado de las frecuencias por cada uno de los valores.
    '''
    y=datos.dropna()
    repeticiones = 0
    x=[]
    rep=[]
    for i in y:
        x.append(str(i))
    for i in x:
        n = x.count(i)
        if n > repeticiones:
            repeticiones = n
            rep.append(repeticiones)
    return(max(rep))
```

```python
def moda(datos):
    '''
    Función moda
    Toma cada una de las variables que se encuentran en el base, para realizar el cálculo de la moda
    moda: espacio donde se guardará el listado de los valores de mayor frecuencia para cada variable.
    n: espacio donde queda la cantidad de veces que aparece cada elemento de la lista
    '''
    repeticiones = 0
    y=datos.dropna()
    x=[]
    for j in y:
        x.append(str(j))
    for j in x:
        n = x.count(j)
        if n > repeticiones:
            repeticiones = n

    sali = [] #Arreglo donde se guardara el o los valores de mayor frecuencia 

    for j in x:
        n = x.count(j) # Devuelve el número de veces que x aparece enla lista.
        if n == repeticiones and j not in sali:
            sali.append(j)
    return(sali)
```

```python
def atributos_data(x):
  '''
  Función atributos_data
  Toma cada una de las variables que se encuentran en el base, para crear un data frame en el cual cada fila corresponde a cada una de las variables
  existentes en la base y cada columna corresponde a cada uno de los siguintes cálculos:
  - Nombre de la columna
  - Tipo de dato de la columna (incluido: int, float, string, datetime, boolean, object)
  - Tipo de la variable (continua, numérica discreta, categórica)
  - Cantidad de valores nulos
  - Valor que más se repite
  - Frecuencia del valor que más se repite
  - Valor máximo (este cálculo aplica unicamente para variables de tipo "int" y "float")
  - Valor mínimo (este cálculo aplica unicamente para variables de tipo "int" y "float")
  '''
  columnas = x.columns.to_list()
  nombre=[]
  tipo_dato=[]
  tipo_variable=[]
  valores_unicos=[]
  valores_nulos=[]
  Mayor_frecuencia=[]
  frecuencia=[]
  Maximo=[]
  Minimo=[]
  raiz=[]
  for i in columnas:
    nombre.append(i)
    if x[i].dtype == 'O':
      dato = 'string'
      variable = 'categórica'
    elif x[i].dtype == 'int64':
      dato = 'int'
      variable = 'discreta'
    elif x[i].dtype == 'float64':
      dato = 'float'
      variable = 'continua'
    elif x[i].dtype == 'datetime64[ns]':
      dato = 'datetime'
      variable = 'categórica (date)'  
    tipo_dato.append(dato)
    tipo_variable.append(variable)
    valores_unicos.append(len(x[i].unique()))
    valores_nulos.append(x[i].isnull().sum())
    Mayor_frecuencia.append(moda(x[i]))
    frecuencia.append(mayor(x[i]))
    y=x[i].dropna()
    if x[i].dtype == 'O':
          Maximo.append("N/A")
    elif x[i].dtype == 'datetime64[ns]':
          Maximo.append("N/A")
    elif x[i].dtype == 'int64':
        Maximo.append(str(max(y)))
    elif x[i].dtype == 'float64':
        Maximo.append(str(max(y)))
    if x[i].dtype == 'O':
          Minimo.append("N/A")
    elif x[i].dtype == 'datetime64[ns]':
          Minimo.append("N/A")
    elif x[i].dtype == 'int64':
        Minimo.append(str(min(y)))
    elif x[i].dtype == 'float64':
        Minimo.append(str(min(y)))
    raiz.append(round(len(x[i])**0.5, 2) > len(x[i].unique()))
  atributos = pd.DataFrame({'Nombre_Columna': nombre, 'Tipo de dato': tipo_dato, 'Tipo de Variable': tipo_variable,
                            'Valores únicos': valores_unicos, 'Valores nulos': valores_nulos, 'dato frecuente': Mayor_frecuencia,
                            'Total dato frec.': frecuencia, 'Valor máximo': Maximo, 'Valor mínimo': Minimo})
  return atributos
```
